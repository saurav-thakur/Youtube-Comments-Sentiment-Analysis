```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Dataset 1

```python
data1 = pd.read_csv("../dataset/sentiment_dataset_four_genres/Education.csv")
data2 = pd.read_csv("../dataset/sentiment_dataset_four_genres/Finance.csv")
data3 = pd.read_csv("../dataset/sentiment_dataset_four_genres/Politics.csv")
data4 = pd.read_csv("../dataset/sentiment_dataset_four_genres/Sports.csv")
```

```python
data4.head()
```

```python
dataset1 = pd.concat([data1,data2,data3,data4],axis="rows")
dataset1.columns = dataset1.columns.str.lower()
dataset1
```

```python
dataset1["label"] = dataset1["label"].replace({
    "negative":0,
    "positive":1,
})
```

Dataset 2

```python
data1 = pd.read_csv("../dataset/twitter_dataset/train.csv",encoding="unicode_escape")
data2 = pd.read_csv("../dataset/twitter_dataset/test.csv",encoding="unicode_escape")
data2
```

```python
data1
```

```python
dataset2 = pd.concat([data1,data2],axis="rows")
dataset2
```

```python
dataset2 = dataset2[["text","sentiment"]]
dataset2.dropna(inplace=True)
dataset2.rename(columns={"sentiment":"label"},inplace=True)
dataset2
```

```python
dataset2["label"] = dataset2["label"].replace({
    "negative":0,
    "positive":1,
    "neutral":2,
})
```

Dataset 3

```python
"""
0 Indicating it is a Neutral Tweet/Comment
2.1 Indicating a Postive Sentiment
3.-1 Indicating a Negative Tweet/Comment

"""

data1 = pd.read_csv("../dataset/twitter_reddit_dataset/Reddit_Data.csv")

data2 = pd.read_csv("../dataset/twitter_reddit_dataset/Twitter_Data.csv")

data1.rename(columns={"clean_comment":"text","category":"label"},inplace=True)
data2.rename(columns={"clean_text":"text","category":"label"},inplace=True)

data1
```

```python
dataset3 = pd.concat([data1,data2],axis="rows")
dataset3.dropna(inplace=True)
dataset3
```

```python
dataset3["label"] = dataset3["label"].replace({
    -1.0:0,
    1.0:1,
})
dataset3["label"] = dataset3["label"].astype(int)
dataset3
```

Dataset 4

```python
data1 = pd.read_csv("../dataset/imdb_dataset/Train.csv")
data2 = pd.read_csv("../dataset/imdb_dataset/Test.csv")
data3 = pd.read_csv("../dataset/imdb_dataset/Valid.csv")
data1
```

```python
# 0 is negative and 1 is positive
dataset4 = pd.concat([data1,data2],axis="rows")
dataset4.dropna(inplace=True)
dataset4
```

```python
final_dataset = pd.concat([dataset1,dataset2,dataset3,dataset4],axis="rows")
final_dataset.shape
```
